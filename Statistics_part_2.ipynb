1.What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to determine the reliability of a study by evaluating a hypothesis (a statement about a population) against data, aiming to either reject or fail to reject the initial assumption.



2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
In hypothesis testing, the null hypothesis (H0) assumes no effect or difference, while the alternative hypothesis (H1 or Ha) proposes an effect or difference exists, and is accepted if evidence rejects the null hypothesis





3. What is the significance level in hypothesis testing, and why is it important

Significance levels and P values are important tools that help you quantify and control this type of error in a hypothesis test. Using these tools to decide when to reject the null hypothesis increases your chance of making the correct decision.


4.What does a P-value represent in hypothesis testing

In hypothesis testing, a p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. It essentially measures the strength of evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing

In hypothesis testing, the p-value indicates the probability of observing results as extreme as, or more extreme than, the ones obtained, assuming the null hypothesis is true. A small p-value (typically less than 0.05) suggests strong evidence against the null hypothesis, leading to its rejection in favor of the alternative hypothesis.

6.What are Type 1 and Type 2 errors in hypothesis testing

In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, while a Type II error (false negative) occurs when you fail to reject a false null hypothesis.

7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

In hypothesis testing, a one-tailed test checks for a difference in one specific direction (e.g., greater than or less than), while a two-tailed test checks for a difference in either direction (greater or less than, or just "different").

8.What is the Z-test, and when is it used in hypothesis testing

A Z-test is a statistical test used to determine if there's a significant difference between a sample mean and a population mean or between the means of two populations when the population standard deviation is known or the sample size is large, typically greater than 30.

9.How do you calculate the Z-score, and what does it represent in hypothesis testing?

To calculate a z-score, you subtract the population mean from a raw score and divide the result by the population standard deviation. In hypothesis testing, a z-score indicates how many standard deviations a data point or sample mean is from the population mean, helping determine if the observed data is likely under the null hypothesis.

You must use the t-distribution table when working problems when the population standard deviation (σ) is not known and the sample size is small (n<30). If σ is not known, then using t-distribution is correct. eIf σ is known, then using the normal distribution is correct.

The primary difference between a Z-test and a T-test lies in when they are used: Z-tests are used for large sample sizes (typically n > 30) and known population standard deviation, while T-tests are used for small sample sizes (typically n < 30) or when the population standard deviation is unknown.


12.What is the T-test, and how is it used in hypothesis testing?

A t-test is a statistical test used to determine if there's a significant difference between the means of two groups, and it's a crucial tool in hypothesis testing to evaluate whether a process or treatment has an effect on a population.

```python
#Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results

import scipy.stats as stats

def z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample_mean: Mean of the sample
    - population_mean: Known population mean (null hypothesis)
    - population_std_dev: Known population standard deviation
    - sample_size: Size of the sample
    - alpha: Significance level (default 0.05)
    - alternative: 'two-sided', 'less', or 'greater'

    Returns:
    - z_score, p_value, interpretation
    """

    se = population_std_dev / (sample_size ** 0.5)

    # Calculate z-score
    z = (sample_mean - population_mean) / se

    # Determine p-value
    if alternative == 'two-sided':
        p = 2 * (1 - stats.norm.cdf(abs(z)))
    elif alternative == 'greater':
        p = 1 - stats.norm.cdf(z)
    elif alternative == 'less':
        p = stats.norm.cdf(z)
    else:
        raise ValueError("Alternative must be 'two-sided', 'greater', or 'less'.")

    # Interpret the result
    if p < alpha:
        result = f"Reject the null hypothesis (p = {p:.4f} < α = {alpha})"
    else:
        result = f"Fail to reject the null hypothesis (p = {p:.4f} ≥ α = {alpha})"

    return z, p, result

# Example usage:
sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 50
alpha = 0.05
alternative = 'greater'  # Test if sample_mean > population_mean

z_score, p_value, interpretation = z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha, alternative)

print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
#Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

# Sample data
np.random.seed(1)
sample = np.random.normal(loc=52, scale=5, size=40)  # sample drawn from N(52, 5)

# Known population parameters
population_mean = 50
population_std = 5

# Sample statistics
sample_mean = np.mean(sample)
n = len(sample)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # two-tailed test

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population means.")
```

```python
# Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=52, scale=5, size=40)

# Known population values
population_mean = 50
population_std = 5
alpha = 0.05  # significance level

# Sample statistics
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Z critical values for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Z-critical: ±{z_critical:.2f}")

# Decision
if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# --- Visualization ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade critical regions
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')

# Plot the Z-score
plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.axvline(z_critical, color='red', linestyle='--')

# Labels and legend
plt.title("Two-Tailed Z-Test: Decision Regions")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=50, mu1=52, sigma=5, n=30, alpha=0.05, test_type='right'):
    """
    Visualizes Type I and Type II errors for a one-tailed Z-test.

    Parameters:
    - mu0: Null hypothesis mean
    - mu1: Alternative hypothesis mean
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    - test_type: 'right' for right-tailed, 'left' for left-tailed, 'two' for two-tailed
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value(s)
    if test_type == 'right':
        z_crit = norm.ppf(1 - alpha)
        x_crit = mu0 + z_crit * se
        beta = norm.cdf((x_crit - mu1) / se)
    elif test_type == 'left':
        z_crit = norm.ppf(alpha)
        x_crit = mu0 + z_crit * se
        beta = 1 - norm.cdf((x_crit - mu1) / se)
    elif test_type == 'two':
        z_crit = norm.ppf(1 - alpha/2)
        x_crit_upper = mu0 + z_crit * se
        x_crit_lower = mu0 - z_crit * se
        beta = norm.cdf((x_crit_upper - mu1)/se) - norm.cdf((x_crit_lower - mu1)/se)
    else:
        raise ValueError("test_type must be 'left', 'right', or 'two'")

    # x-axis range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀ Distribution', color='blue')
    plt.plot(x, y1, label='H₁ Distribution', color='green')

    if test_type == 'right':
        plt.fill_between(x, y0, where=x >= x_crit, color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, y1, where=x < x_crit, color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical value = {x_crit:.2f}')
    elif test_type == 'left':
        plt.fill_between(x, y0, where=x <= x_crit, color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, y1, where=x > x_crit, color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical value = {x_crit:.2f}')
    elif test_type == 'two':
        plt.fill_between(x, y0, where=(x <= x_crit_lower) | (x >= x_crit_upper), color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, y1, where=(x > x_crit_lower) & (x < x_crit_upper), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(x_crit_lower, color='black', linestyle='--')
        plt.axvline(x_crit_upper, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")
```

```python
#Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy import stats

# Simulate two independent samples
np.random.seed(123)
group1 = np.random.normal(loc=100, scale=15, size=30)  # Group 1: mean = 100
group2 = np.random.normal(loc=110, scale=15, size=30)  # Group 2: mean = 110

# Perform Independent T-test
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)

# Print statistics
print("Group 1 Mean:", np.mean(group1))
print("Group 2 Mean:", np.mean(group2))
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpret results
alpha = 0.05
print("\n--- Hypothesis Test Result ---")
print("Null Hypothesis: The two group means are equal.")
print("Alternative Hypothesis: The two group means are different.")

if p_value < alpha:
    print(f"Since p-value ({p_value:.4f}) < α ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the group means.")
else:
    print(f"Since p-value ({p_value:.4f}) >= α ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: No statistically significant difference between the group means.")
```

```python
# Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate paired data (e.g., before and after treatment)
np.random.seed(42)
n = 30
before = np.random.normal(loc=70, scale=10, size=n)
after = before + np.random.normal(loc=-5, scale=5, size=n)  # some improvement

# Perform paired sample T-test
t_stat, p_value = stats.ttest_rel(before, after)

# Print results
print("Mean Before:", np.mean(before))
print("Mean After :", np.mean(after))
print("T-statistic :", t_stat)
print("P-value     :", p_value)

# Interpretation
alpha = 0.05
print("\n--- Hypothesis Test Result ---")
print("Null Hypothesis: Mean difference = 0 (no change).")
print("Alternative Hypothesis: Mean difference ≠ 0.")

if p_value < alpha:
    print(f"Since p-value ({p_value:.4f}) < α ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant change.")
else:
    print(f"Since p-value ({p_value:.4f}) ≥ α ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: No statistically significant change.")

# --- Visualization ---
plt.figure(figsize=(10, 6))
plt.plot(before, label='Before', marker='o')
plt.plot(after, label='After', marker='o')
plt.title('Before vs After (Paired Sample)')
plt.xlabel('Subject Index')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - Tuple (lower bound, upper bound) of the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(n)  # standard error of the mean
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * sem
    ci = (mean - margin_of_error, mean + margin_of_error)

    print(f"Sample Mean        : {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
    return ci
```

```python
#Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level and sample data.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - Margin of error (float)
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)
    sem = sample_std / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_crit * sem

    print(f"Sample Size (n): {n}")
    print(f"Sample Mean    : {np.mean(data):.2f}")
    print(f"Sample Std Dev : {sample_std:.2f}")
    print(f"Confidence Level: {int(confidence*100)}%")
    print(f"Margin of Error: ±{moe:.2f}")

    return moe
```

```python
#Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, false_positive, prevalence_complement=None):
    """
    Computes the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H), the prior probability (e.g., P(Disease) = 0.01)
    - likelihood: P(E|H), probability of evidence given hypothesis (e.g., test is positive if you have disease)
    - false_positive: P(E|~H), probability of evidence given no hypothesis (false positive rate)
    - prevalence_complement: Optional. P(~H), prior probability of not having the disease. If None, it's 1 - prior.

    Returns:
    - posterior: P(H|E)
    """
    if prevalence_complement is None:
        prevalence_complement = 1 - prior

    numerator = likelihood * prior
    denominator = (likelihood * prior) + (false_positive * prevalence_complement)

    posterior = numerator / denominator
    return posterior

# Example values
prior = 0.01               # P(Disease)
likelihood = 0.99          # P(Positive | Disease)
false_positive = 0.01      # P(Positive | No Disease)

posterior = bayes_theorem(prior, likelihood, false_positive)
print(f"Probability of actually having the disease given a positive test: {posterior:.4f}")
```

```python
#Perform a Chi-square test for independence between two categorical variables in Python

import pandas as pd
import scipy.stats as stats

# Step 1: Create the contingency table
data = [[30, 10],   # Male
        [20, 40]]   # Female

# Optional: for readability
index = ['Male', 'Female']
columns = ['Coffee', 'Tea']

contingency_table = pd.DataFrame(data, index=index, columns=columns)
print("Contingency Table:\n", contingency_table)

# Step 2: Perform Chi-square test of independence
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Step 3: Display results
print("\nChi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:\n", pd.DataFrame(expected, index=index, columns=columns))

# Step 4: Interpret result
alpha = 0.05
if p < alpha:
    print("\nResult: Reject the null hypothesis. There is a significant association between gender and drink preference.")
else:
    print("\nResult: Fail to reject the null hypothesis. No significant association between gender and drink preference.")

```

```python
#Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test of independence.

    Parameters:
    - observed: a 2D list or DataFrame of observed counts

    Returns:
    - expected: a DataFrame of expected frequencies
    """
    if not isinstance(observed, pd.DataFrame):
        observed = pd.DataFrame(observed)

    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.values.sum()

    expected = pd.DataFrame(index=observed.index, columns=observed.columns)

    for row in observed.index:
        for col in observed.columns:
            expected.loc[row, col] = (row_totals[row] * col_totals[col]) / grand_total

    return expected.astype(float)

# Example observed data
observed_data = pd.DataFrame(
    [[30, 10],   # Male
     [20, 40]],  # Female
    index=["Male", "Female"],
    columns=["Coffee", "Tea"]
)

print("Observed Frequencies:")
print(observed_data)

expected_data = calculate_expected_frequencies(observed_data)

print("\nExpected Frequencies:")
print(expected_data)
```

```python
#Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import scipy.stats as stats
import numpy as np

# Observed frequencies from rolling a die 60 times
observed = np.array([8, 9, 10, 12, 11, 10])

# Expected frequencies for a fair die
expected = np.array([10, 10, 10, 10, 10, 10])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The die may not be fair.")
else:
    print("Result: Fail to reject the null hypothesis. No evidence the die is unfair.")
```

```python
#Implement an F-test using Python to compare the variances of two random samples

import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1, sample2: Arrays/lists of data values
    - alpha: Significance level (default 0.05)

    Returns:
    - F-statistic, p-value, interpretation
    """

    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Determine which is larger
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    # Two-tailed p-value
    p_value = 2 * min(
        stats.f.cdf(f_stat, dfn, dfd),
        1 - stats.f.cdf(f_stat, dfn, dfd)
    )

    # Interpretation
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p = {p_value:.4f} < α = {alpha}). Variances are significantly different."
    else:
        interpretation = f"Fail to reject the null hypothesis (p = {p_value:.4f} ≥ α = {alpha}). No significant difference in variances."

    return f_stat, p_value, interpretation

# Example usage
np.random.seed(42)
sample1 = np.random.normal(10, 5, 30)   # mean=10, std=5
sample2 = np.random.normal(10, 2, 30)   # mean=10, std=2

f_stat, p_val, result = f_test(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Interpretation: {result}")
```

```python
#Write a Python program to perform an ANOVA test to compare means between multiple groups and
#interpret the results

import numpy as np
import scipy.stats as stats

# Step 1: Create sample data for multiple groups
# For reproducibility
np.random.seed(42)

# Simulate data from 3 groups
group1 = np.random.normal(20, 5, 30)   # mean=20, std=5
group2 = np.random.normal(22, 5, 30)   # mean=22, std=5
group3 = np.random.normal(27, 5, 30)   # mean=27, std=5

# Step 2: Perform One-way ANOVA test
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Step 3: Display results
print("One-way ANOVA Test Results:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. At least one group mean is significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference between group means.")
```

```python
#Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups, alpha=0.05):
    """
    Check normality and equal variance assumptions for ANOVA.

    Parameters:
    - groups: list of arrays/lists representing each group
    - alpha: significance level for tests

    Returns:
    - Prints test results and visualizations
    """
    print("=== Checking ANOVA Assumptions ===")

    # Flatten groups for variance test
    all_data = np.concatenate(groups)
    group_labels = np.concatenate([[f'Group {i+1}'] * len(group) for i, group in enumerate(groups)])

    # 1. Normality check (Shapiro-Wilk for each group)
    print("\nNormality (Shapiro-Wilk Test):")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        result = "Pass" if p > alpha else "Fail"
        print(f"  Group {i+1}: W={stat:.4f}, p={p:.4f} → {result}")

    # 2. Equal variance check (Levene’s test)
    print("\nEqual Variance (Levene's Test):")
    levene_stat, levene_p = stats.levene(*groups)
    levene_result = "Pass" if levene_p > alpha else "Fail"
    print(f"  Levene’s test: W={levene_stat:.4f}, p={levene_p:.4f} → {levene_result}")

    # 3. Q-Q plots for each group (visual normality check)
    print("\nGenerating Q-Q plots for visual normality check...")
    fig, axes = plt.subplots(1, len(groups), figsize=(5 * len(groups), 4))
    for i, group in enumerate(groups):
        stats.probplot(group, dist="norm", plot=axes[i])
        axes[i].set_title(f"Q-Q Plot: Group {i+1}")
    plt.tight_layout()
    plt.show()

    # 4. Boxplot to visualize variances
    print("Generating boxplot to visualize variance...")
    sns.boxplot(x=group_labels, y=all_data)
    plt.title("Group Variance Visualization")
    plt.xlabel("Group")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()
```

```python
np.random.seed(42)

group1 = np.random.normal(20, 5, 30)
group2 = np.random.normal(21, 5, 30)
group3 = np.random.normal(22, 5, 30)

check_anova_assumptions([group1, group2, group3])

```

```python
#Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
#results

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Simulate data
np.random.seed(42)
n = 10  # samples per group

# Factors
treatments = ['A', 'B']
genders = ['Male', 'Female']

# Create dataset
data = []

for treatment in treatments:
    for gender in genders:
        base = 50 if treatment == 'A' else 55
        gender_effect = 5 if gender == 'Female' else 0
        values = base + gender_effect + np.random.normal(0, 3, n)
        for value in values:
            data.append([treatment, gender, value])

df = pd.DataFrame(data, columns=['Treatment', 'Gender', 'Score'])

# Step 2: Two-way ANOVA
model = ols('Score ~ C(Treatment) + C(Gender) + C(Treatment):C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("=== Two-Way ANOVA Table ===")
print(anova_table)

# Step 3: Visualization - Interaction Plot
plt.figure(figsize=(8, 5))
sns.pointplot(data=df, x='Treatment', y='Score', hue='Gender',
              dodge=True, markers=["o", "s"], capsize=.1, errwidth=1, palette="Set2")
plt.title("Interaction Plot: Treatment × Gender")
plt.ylabel("Mean Score")
plt.grid(True)
plt.show()
```

```python
#Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(df1=5, df2=20, alpha=0.05):
    """
    Plot the F-distribution with given degrees of freedom and significance level.
    Highlights the critical region for a one-tailed test.
    """
    # Create a range of F values
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    # Critical value for right tail
    critical_value = stats.f.ppf(1 - alpha, df1, df2)

    # Plot the F-distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.3, label='Critical Region')

    # Add lines and labels
    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical value (α={alpha}) = {critical_value:.2f}')
    plt.title('F-distribution and Critical Region')
    plt.xlabel('F-statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print critical value
    print(f"Critical value (F({df1}, {df2}), α={alpha}): {critical_value:.4f}")

# Example usage
plot_f_distribution(df1=5, df2=20, alpha=0.05)
```

```python
#Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Simulate sample data
np.random.seed(42)

group1 = np.random.normal(70, 5, 30)   # Group 1: mean=70
group2 = np.random.normal(75, 5, 30)   # Group 2: mean=75
group3 = np.random.normal(80, 5, 30)   # Group 3: mean=80

# Step 2: Combine data into a DataFrame
df = pd.DataFrame({
    'Score': np.concatenate([group1, group2, group3]),
    'Group': ['Group 1'] * 30 + ['Group 2'] * 30 + ['Group 3'] * 30
})

# Step 3: Perform one-way ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Step 4: Display results
print("=== One-Way ANOVA Test ===")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — at least one group mean is significantly different.")
else:
    print("Conclusion: Fail to reject H₀ — no significant difference between group means.")

# Step 5: Visualize with boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Group', y='Score', palette='Set3')
plt.title('One-Way ANOVA: Group Mean Comparison')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
#Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate sample data
np.random.seed(1)
sample = np.random.normal(loc=52, scale=5, size=40)  # mean=52, std=5, n=40

# Step 2: Define the null hypothesis value (population mean)
mu = 50

# Step 3: Perform one-sample t-test
t_stat, p_val = stats.ttest_1samp(sample, mu)

# Step 4: Output the results
print("=== One-Sample t-Test ===")
print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject H₀ → Sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject H₀ → No significant difference from the population mean.")

# Step 5: Visualize the distribution
plt.hist(sample, bins=10, color='skyblue', edgecolor='black')
plt.axvline(np.mean(sample), color='red', linestyle='--', label=f"Sample Mean = {np.mean(sample):.2f}")
plt.axvline(mu, color='green', linestyle='--', label=f"Population Mean = {mu}")
plt.title("Simulated Sample Data (Normal Distribution)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=5, size=30)  # true std = 5, so true variance = 25

# Step 2: Set the hypothesized population variance
sigma_0_squared = 25  # H0: population variance = 25

# Step 3: Calculate sample variance and test statistic
n = len(sample)
sample_variance = np.var(sample, ddof=1)  # sample variance (unbiased)
chi2_stat = (n - 1) * sample_variance / sigma_0_squared

# Step 4: Compute critical values and p-value (two-tailed test)
alpha = 0.05
df = n - 1
chi2_lower = stats.chi2.ppf(alpha / 2, df)
chi2_upper = stats.chi2.ppf(1 - alpha / 2, df)
p_value = 2 * min(stats.chi2.cdf(chi2_stat, df), 1 - stats.chi2.cdf(chi2_stat, df))

# Step 5: Output results
print("=== Chi-Square Test for Population Variance ===")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical Chi2 values (α = {alpha}): [{chi2_lower:.4f}, {chi2_upper:.4f}]")
print(f"P-value: {p_value:.4f}")

if chi2_stat < chi2_lower or chi2_stat > chi2_upper:
    print("Conclusion: Reject H₀ → The population variance is significantly different from", sigma_0_squared)
else:
    print("Conclusion: Fail to reject H₀ → No significant evidence the variance differs from", sigma_0_squared)
```

```python
#Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Step 1: Define success counts and sample sizes
success_a = 56
nobs_a = 100

success_b = 42
nobs_b = 100

# Step 2: Perform two-proportion z-test
count = np.array([success_a, success_b])
nobs = np.array([nobs_a, nobs_b])

z_stat, p_val = proportions_ztest(count, nobs)

# Step 3: Output the results
print("=== Two-Proportion Z-Test ===")
print(f"Proportion A: {success_a/nobs_a:.2f}")
print(f"Proportion B: {success_b/nobs_b:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject H₀ → There is a significant difference in proportions.")
else:
    print("Conclusion: Fail to reject H₀ → No significant difference in proportions.")
```

```python
#Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=30)  # Variance ~ 100
data2 = np.random.normal(loc=55, scale=5, size=30)   # Variance ~ 25

# Step 2: Calculate variances and F-statistic
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
f_stat = var1 / var2

df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 3: Calculate p-value (two-tailed test)
p_val = 2 * min(
    stats.f.cdf(f_stat, df1, df2),
    1 - stats.f.cdf(f_stat, df1, df2)
)

# Step 4: Display results
print("=== F-Test for Equality of Variances ===")
print(f"Variance of Group 1: {var1:.2f}")
print(f"Variance of Group 2: {var2:.2f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"Degrees of Freedom: df1 = {df1}, df2 = {df2}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject H₀ → The variances are significantly different.")
else:
    print("Conclusion: Fail to reject H₀ → No significant difference in variances.")

# Step 5: Visualization
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=10, alpha=0.6, label='Group 1', color='skyblue', edgecolor='black')
plt.hist(data2, bins=10, alpha=0.6, label='Group 2', color='salmon', edgecolor='black')
plt.title('Distribution of Two Groups')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
